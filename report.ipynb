```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n_hidden_layer = 1
data_filename = "report_nh_" + str(n_hidden_layer) + ".csv"
df = pd.read_csv("report_nh_1.csv", header=None)
df.columns = ['folds_list', 'cv_test_err_list', 'cv_train_err_list', 'cv_test_precision_1', 'cv_test_precision_2', 'cv_test_precision_3', 'cv_test_recall_1', 'cv_test_recall_2', 'cv_test_recall_3', 'cv_test_f1_score_1', 'cv_test_f1_score_2', 'cv_test_f1_score_3', 'cv_train_precision_1', 'cv_train_precision_2', 'cv_train_precision_3', 'cv_train_recall_1', 'cv_train_recall_2', 'cv_train_recall_3', 'cv_train_f1_score_1', 'cv_train_f1_score_2', 'cv_train_f1_score_3']
df.head()
```

```python
iris_labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

def plt_mse(df, n_hidden_layer):
    plt.figure(figsize=(12, 8))
    plt.plot(df['folds_list'], df['cv_test_err_list'], label='test')
    plt.plot(df['folds_list'], df['cv_train_err_list'], label='train')
    plt.xlabel('Epochs')
    plt.ylabel('Mean square error')
    plt.legend()
    filename = 'N_LAYER_' + str(n_hidden_layer) + '_mse_train_test.png'
    plt.savefig(filename)
    plt.show()
    
def plt_classification(df, train_or_test, precision_or_recall_or_f1, n_hidden_layer):
    plt.figure(figsize=(12, 8))
    x = df['folds_list']
    
    for i in range(len(iris_labels)):
        y = df['cv_' + train_or_test + '_' + precision_or_recall_or_f1 + '_' + str(i + 1)]
        plt.plot(x, y, label=iris_labels[i])
    
    plt.xlabel('Epochs')
    plt.ylabel(precision_or_recall_or_f1)
    plt.legend()
    plt.title(train_or_test.upper() + ': ' + 'EPOCH vs ' + precision_or_recall_or_f1.upper())
    filename = 'N_LAYER_' + str(n_hidden_layer) + '_' + train_or_test + '_' + precision_or_recall_or_f1 + '.png'
    plt.savefig(filename)
    plt.show()
```

```python
plt_mse(df, n_hidden_layer=n_hidden_layer)

plt_classification(df, train_or_test='train', precision_or_recall_or_f1='precision', n_hidden_layer=n_hidden_layer)
plt_classification(df, train_or_test='train', precision_or_recall_or_f1='recall', n_hidden_layer=n_hidden_layer)
plt_classification(df, train_or_test='train', precision_or_recall_or_f1='f1_score', n_hidden_layer=n_hidden_layer)

plt_classification(df, train_or_test='test', precision_or_recall_or_f1='precision', n_hidden_layer=n_hidden_layer)
plt_classification(df, train_or_test='test', precision_or_recall_or_f1='recall', n_hidden_layer=n_hidden_layer)
plt_classification(df, train_or_test='test', precision_or_recall_or_f1='f1_score', n_hidden_layer=n_hidden_layer)
```
